```python



import matplotlib.pyplot as plt
from scipy.stats import shapiro

# Function to plot histogram and perform Shapiro-Wilk test
def analyze_distribution(data, column):
    plt.hist(data[column], bins=30, alpha=0.75)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
    
    stat, p = shapiro(data[column].dropna())  # dropna() to remove missing values for the test
    print(f'Shapiro-Wilk Test for {column}: Statistics={stat:.3f}, p={p:.3f}')
    if p > 0.05:
        print('Data looks Gaussian (fail to reject H0)')
    else:
        print('Data does not look Gaussian (reject H0)')

analyze_distribution(data_togo, 'GHI')  
```

The histograms and summary statistics for the GHI (Global Horizontal Irradiance) from Togo, Benin, and Sierra Leone provide insightful initial data on the solar energy potential in each location. Here’s a summary of the key observations:

### Histogram Observations:
- Togo (Dapaong): The GHI distribution shows both negative values (possibly nighttime readings or sensor errors) and a wide range of positive values, indicating variable solar exposure.
- Benin (Malanville): Similar to Togo, with a spread of negative and positive GHI values. The histogram suggests a slightly more concentrated distribution of higher GHI values compared to Togo.
- Sierra Leone (Bumbuna): Shows a narrower spread of GHI values, mostly concentrated around lower positive values, indicating potentially less solar irradiance than the other two locations.

### Summary Statistics Key Points:
- Mean GHI:
  - Togo: 230.56 W/m²
  - Benin: 226.52 W/m²
  - Sierra Leone: 198.24 W/m²
- Standard Deviation:
  - Togo: 322.53 W/m²
  - Benin: 319.45 W/m²
  - Sierra Leone: 288.89 W/m²
- Maximum GHI:
  - Togo: 1424.0 W/m²
  - Benin: 1417.0 W/m²
  - Sierra Leone: 1473.0 W/m²

### Initial Analysis and Further Steps:
- Potential: Benin and Togo show similar potential with a slightly higher maximum GHI in Sierra Leone, though Sierra Leone's average GHI is lower. The variability (standard deviation) in GHI is also slightly lower in Sierra Leone, suggesting more consistent but lower solar irradiance.
- Data Quality: The presence of negative GHI values should be further investigated or cleaned, as these likely represent erroneous or night-time readings.
- Next Steps in EDA:
  - Time Series Analysis: To better understand how GHI, DNI, and DHI vary with time across different times of the day and seasons.
  - Correlation Analysis: Examine how GHI correlates with other environmental variables (like temperature and humidity) which might impact solar panel efficiency.

```python
# Convert 'Timestamp' from string to datetime for plotting
data_togo['Timestamp'] = pd.to_datetime(data_togo['Timestamp'])
data_benin['Timestamp'] = pd.to_datetime(data_benin['Timestamp'])
data_sierra_leone['Timestamp'] = pd.to_datetime(data_sierra_leone['Timestamp'])

# Creating time series plots for GHI
fig, ax = plt.subplots(3, 1, figsize=(15, 18))

ax[0].plot(data_togo['Timestamp'], data_togo['GHI'], label='GHI - Togo', color='blue')
ax[0].set_title('Time Series of GHI - Togo (Dapaong)')
ax[0].set_ylabel('GHI (W/m²)')
ax[0].legend()

ax[1].plot(data_benin['Timestamp'], data_benin['GHI'], label='GHI - Benin', color='green')
ax[1].set_title('Time Series of GHI - Benin (Malanville)')
ax[1].set_ylabel('GHI (W/m²)')
ax[1].legend()

ax[2].plot(data_sierra_leone['Timestamp'], data_sierra_leone['GHI'], label='GHI - Sierra Leone', color='red')
ax[2].set_title('Time Series of GHI - Sierra Leone (Bumbuna)')
ax[2].set_ylabel('GHI (W/m²)')
ax[2].legend()

plt.tight_layout()
plt.show()
```
