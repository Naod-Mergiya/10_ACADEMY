```python
import matplotlib.pyplot as plt

data_togo['GHI'].hist(bins=50)
plt.title('Distribution of GHI in Togo-Dapaong')
plt.xlabel('GHI')
plt.ylabel('Frequency')
plt.show()
```

```python
data_togo[['GHI', 'DNI', 'DHI']].plot(kind='box', vert=False)
plt.title('Box Plot of Solar Radiation Measurements in Togo-Dapaong')
plt.show()
```

```python
import seaborn as sns

# Correlation matrix heatmap
# Drop non-numeric columns
numeric_data_togo = data_togo.select_dtypes(include='number')

# Correlation matrix heatmap
correlation_matrix = numeric_data_togo.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for Togo-Dapaong')
plt.show()
```

```python
import seaborn as sns

# Setting up visualization aesthetics
sns.set(style="whitegrid")

# Create summary statistics for each dataset
summary_togo = data_togo.describe()
summary_benin = data_benin.describe()
summary_sierraleone = data_sierra_leone.describe()

# Plotting histograms for GHI for each region to understand distribution of solar radiation
fig, ax = plt.subplots(3, 1, figsize=(12, 18))

ax[0].hist(data_togo['GHI'], bins=50, color='skyblue', edgecolor='black')
ax[0].set_title('Histogram of GHI - Togo (Dapaong)')
ax[0].set_xlabel('GHI (W/m)')
ax[0].set_ylabel('Frequency')

ax[1].hist(data_benin['GHI'], bins=50, color='lightgreen', edgecolor='black')
ax[1].set_title('Histogram of GHI - Benin (Malanville)')
ax[1].set_xlabel('GHI (W/m)')
ax[1].set_ylabel('Frequency')

ax[2].hist(data_sierra_leone['GHI'], bins=50, color='salmon', edgecolor='black')
ax[2].set_title('Histogram of GHI - Sierra Leone (Bumbuna)')
ax[2].set_xlabel('GHI (W/m)')
ax[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()

# Displaying the summary statistics to accompany the histograms
summary_togo, summary_benin, summary_sierraleone
```
