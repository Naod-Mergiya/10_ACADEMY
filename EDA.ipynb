```python
# Checking Distribution Type


import matplotlib.pyplot as plt
from scipy.stats import shapiro

# Function to plot histogram and perform Shapiro-Wilk test
def analyze_distribution(data, column):
    plt.hist(data[column], bins=30, alpha=0.75)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
    
    stat, p = shapiro(data[column].dropna())  # dropna() to remove missing values for the test
    print(f'Shapiro-Wilk Test for {column}: Statistics={stat:.3f}, p={p:.3f}')
    if p > 0.05:
        print('Data looks Gaussian (fail to reject H0)')
    else:
        print('Data does not look Gaussian (reject H0)')

analyze_distribution(data_togo, 'GHI')
analyze_distribution(data_benin, 'GHI')
analyze_distribution(data_sierra_leone, 'GHI')
  
```

```python
# Convert 'Timestamp' from string to datetime for plotting
data_togo['Timestamp'] = pd.to_datetime(data_togo['Timestamp'])
data_benin['Timestamp'] = pd.to_datetime(data_benin['Timestamp'])
data_sierra_leone['Timestamp'] = pd.to_datetime(data_sierra_leone['Timestamp'])

# Creating time series plots for GHI
fig, ax = plt.subplots(3, 1, figsize=(15, 18))

ax[0].plot(data_togo['Timestamp'], data_togo['GHI'], label='GHI - Togo', color='blue')
ax[0].set_title('Time Series of GHI - Togo (Dapaong)')
ax[0].set_ylabel('GHI (W/m²)')
ax[0].legend()

ax[1].plot(data_benin['Timestamp'], data_benin['GHI'], label='GHI - Benin', color='green')
ax[1].set_title('Time Series of GHI - Benin (Malanville)')
ax[1].set_ylabel('GHI (W/m²)')
ax[1].legend()

ax[2].plot(data_sierra_leone['Timestamp'], data_sierra_leone['GHI'], label='GHI - Sierra Leone', color='red')
ax[2].set_title('Time Series of GHI - Sierra Leone (Bumbuna)')
ax[2].set_ylabel('GHI (W/m²)')
ax[2].legend()

plt.tight_layout()
plt.show()
```

#### 9. Wind Analysis
Given the importance of wind data in assessing potential mechanical stresses on solar installations as well as its cooling effect, perform a detailed analysis

```python
# Plot wind speed over time to assess variability and peak gusts
plt.figure(figsize=(14, 7))
plt.plot(data_togo['Timestamp'], data_togo['WS'], label='Wind Speed')
plt.plot(data_togo['Timestamp'], data_togo['WSgust'], label='Wind Gusts', linestyle='--')
plt.title('Wind Speed and Gusts Over Time - Togo')
plt.xlabel('Timestamp')
plt.ylabel('Wind Speed (m/s)')
plt.legend()
plt.show()
```

**Insight:**
Our analysis of wind speed and gusts over time for Togo reveals an intriguing pattern: wind speed tends to be lowest during the dry season, typically spanning from November to March. This observation contrasts with conventional expectations, as the dry season is often associated with increased wind activity due to the influence of the Harmattan winds. However, our findings suggest a unique atmospheric phenomenon or local climatic conditions in Togo, leading to reduced wind speeds during this period.

Understanding this unexpected trend in wind behavior is crucial for MoonLight Energy Solutions' strategic planning. The identification of low wind speeds during the dry season highlights potential challenges and opportunities for renewable energy generation in Togo. While it may pose obstacles for wind energy projects relying solely on wind turbines, it also underscores the importance of diversifying renewable energy sources and exploring alternative solutions such as solar or hybrid systems.

By acknowledging and addressing these nuances in wind dynamics, MoonLight Energy Solutions can develop more resilient and adaptable energy strategies tailored to the specific environmental conditions of Togo. This insight underscores the importance of comprehensive data analysis and localized approaches in driving sustainable energy development initiatives, ultimately contributing to the long-term resilience and energy security of Togo.
